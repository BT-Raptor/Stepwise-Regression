Load the libaries & Data

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
np.set_printoptions(suppress=True)

all_data=pd.read_csv('opp.csv')
all_data.head()
```

```python
all_data=all_data.dropna()
all_data['hdlngth_2']=all_data['hdlngth']**2
all_data['skullw_2']=all_data['skullw']**2
all_data['totlngth_2']=all_data['totlngth']**2
all_data['taill_2']=all_data['taill']**2
all_data['footlgth_2']=all_data['footlgth']**2
all_data['earconch_2']=all_data['earconch']**2
all_data['eye_2']=all_data['eye']**2
all_data['chest_2']=all_data['chest']**2
all_data['belly_2']=all_data['belly']**2

all_data['sex_m']=np.where(all_data['sex']=='m',1,0)
all_data['Pop_Vic']=np.where(all_data['Pop']=='Vic',1,0)
all_data['site_1']=np.where(all_data['site']==1,1,0)
all_data['site_2']=np.where(all_data['site']==2,1,0)
all_data['site_3']=np.where(all_data['site']==3,1,0)
all_data['site_4']=np.where(all_data['site']==4,1,0)
all_data['site_5']=np.where(all_data['site']==5,1,0)
all_data['site_6']=np.where(all_data['site']==6,1,0)
all_data['site_7']=np.where(all_data['site']==7,1,0)

all_data=all_data.drop(['case','site','Pop','sex'],axis=1)
all_data.head()
```

```python
Y=all_data['age']
X=all_data.drop(['age'],axis=1)
model=LinearRegression()
sfs=SequentialFeatureSelector(model,direction='backward')
sfs.fit(X,Y)
sfs.get_feature_names_out()
```

```python
X_vals=sfs.transform(X)
print(X_vals)
```
